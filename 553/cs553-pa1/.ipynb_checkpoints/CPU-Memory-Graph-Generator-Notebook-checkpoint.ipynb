```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline

num_threads = [1, 2, 4, 8]
```

## CPU

```python
### GFlops
flops = pd.DataFrame(num_threads, columns=["Threads"])
flops["GFlops"] = pd.read_csv('cpu/log/flops.log', delim_whitespace=True, usecols=[1], names=["GFlops"])
```

```python
### GIops
iops = pd.DataFrame(num_threads, columns=["Threads"])
iops["GIops"] = pd.read_csv('cpu/log/iops.log', delim_whitespace=True, usecols=[1], names=["GIops"])
```

```python
ops = pd.DataFrame(num_threads, columns=["Threads"])
ops["Floating point"] = flops["GFlops"]
ops["Integer"] = iops["GIops"]
ops
```

```python
p = ops.plot(x="Threads", xticks=num_threads, title="CPU Operations per Second vs. Number of Threads")
p.set_ylabel("Billions of Operations / second (GOps)")
```

## Memory

### Sequential read + write (memcpy)

```python
#### 8 Byte block size
cpy_8b = pd.DataFrame(num_threads, columns=["Threads"])
cpy_8b["MBps"] = pd.read_csv('memory/log/read_and_write/blk-8b.log', delim_whitespace=True, usecols=[1], names=["MBps"])

latency_cpy = pd.DataFrame(num_threads, columns=["Threads"])

## Dividing this by 8 seems to be too small.

## Unit conversion is from Megabytes per second to microseconds per byte
## Therefore can just use the inverse operation (1/x)
latency_cpy["Latency"] = cpy_8b["MBps"] ** -1 / 8
latency_cpy
```

```python
p = latency_cpy.plot(x="Threads", xticks=num_threads, title="Latency for Read+Write operations on 8 B Blocks vs. Number of Threads")
p.set_ylabel("Latency (microseconds)")
```

```python
#### 8 KB block size
cpy_8kb = pd.DataFrame(num_threads, columns=["Threads"])
cpy_8kb["MBps"] = pd.read_csv('memory/log/read_and_write/blk-8000b.log', delim_whitespace=True, usecols=[1], names=["MBps"])
```

```python
#### 8 MB block size
cpy_8mb = pd.DataFrame(num_threads, columns=["Threads"])
cpy_8mb["MBps"] = pd.read_csv('memory/log/read_and_write/blk-8000000b.log', delim_whitespace=True, usecols=[1], names=["MBps"])
```

```python
#### 80 MB block size
cpy_80mb = pd.DataFrame(num_threads, columns=["Threads"])
cpy_80mb["MBps"] = pd.read_csv('memory/log/read_and_write/blk-80000000b.log', delim_whitespace=True, usecols=[1], names=["MBps"])
```

```python
## Now combine all 4 dataframes into one for better visualization

cpy = pd.DataFrame(num_threads, columns=["Threads"])
cpy["8 B Blocks"] = cpy_8b["MBps"]
cpy["8 KB Blocks"] = cpy_8kb["MBps"]
cpy["8 MB Blocks"] = cpy_8mb["MBps"]
cpy["80 MB Blocks"] = cpy_80mb["MBps"]
cpy
```

```python
p = cpy.plot(x="Threads", title="Memory Throughput for Read+Write on Various Block Sizes vs. Number of Threads",xticks=num_threads)
p.set_ylabel("Memory Throughput (Megabytes / second)")
```

### Sequential write access (memset)

```python
#### 8 Byte block size
sw_8b = pd.DataFrame(num_threads, columns=["Threads"])
sw_8b["MBps"] = pd.read_csv('memory/log/seq_write_access/blk-8b.log', delim_whitespace=True, usecols=[1], names=["MBps"])

# Also create a dataframe containing latency values

latency_sw = pd.DataFrame(num_threads, columns=["Threads"])

# Convert MBps (throughput) to latency (in microseconds)
# Multiply by 8 because there 
latency_sw["Latency"] = sw_8b["MBps"] ** -1 / 8
latency_sw
                          
```

```python
p = latency_sw.plot(x="Threads", xticks=num_threads, title="Latency for Sequential Write Operations on 8 B Blocks")
p.set_ylabel("Latency (microseconds)")
```

```python
#### 8 KB block size
sw_8kb = pd.DataFrame(num_threads, columns=["Threads"])
sw_8kb["MBps"] = pd.read_csv('memory/log/seq_write_access/blk-8000b.log', delim_whitespace=True, usecols=[1], names=["MBps"])
```

```python
#### 8 MB block size
sw_8mb = pd.DataFrame(num_threads, columns=["Threads"])
sw_8mb["MBps"] = pd.read_csv('memory/log/seq_write_access/blk-8000000b.log', delim_whitespace=True, usecols=[1], names=["MBps"])
```

```python
#### 80 MB block size
sw_80mb = pd.DataFrame(num_threads, columns=["Threads"])
sw_80mb["MBps"] = pd.read_csv('memory/log/seq_write_access/blk-80000000b.log', delim_whitespace=True, usecols=[1], names=["MBps"])
```

```python
sw = pd.DataFrame(num_threads, columns=["Threads"])
sw["8 B Blocks"] = sw_8b["MBps"]
sw["8 KB Blocks"] = sw_8kb["MBps"]
sw["8 MB Blocks"] = sw_8mb["MBps"]
sw["80 MB Blocks"] = sw_80mb["MBps"]
sw
```

```python
p = sw.plot(x="Threads", xticks=num_threads, title="Memory Throughput for Sequential Write Operations on Various Block Sizes vs. Number of Threads")
p.set_ylabel("Memory Throughput (Megabytes / second)")
```
